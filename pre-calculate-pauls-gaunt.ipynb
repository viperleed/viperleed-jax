```python
from sympy.physics.wigner import gaunt
import numpy as np
```

First calculate all indices and the size of m2 vectors

```python
my_dict = {}
LMAX = 18
for l1 in range(LMAX+1):
    for l2 in range(LMAX+1):
        for l3 in range(abs(l1-l2), l1+l2+1):
            for m1 in range(-l1, l1+1):
                for m2 in range(-l2, l2+1):
                    m3 = -m1-m2
                    if abs(m3)<=l3:
                        l_unsorted = np.array([l1,l2,l3])
                        m_unsorted = np.array([m1,m2,m3])
                        sorted_indices = np.argsort(l_unsorted)[::-1]
                        l = l_unsorted[sorted_indices]
                        m = m_unsorted[sorted_indices]
                        if m[2] < 0:
                            m = -m
                        index = l[0]*(6+l[0]*(11+l[0]*(6+l[0])))/24 + l[1]*(2+l[1]*(3+l[1]))/6 + l[2]*(l[2]+1)/2 + m[2] + 1
                        if index not in my_dict:
                            my_dict[index] = abs(m[1])
                        else:
                            if my_dict[index] < abs(m[1]):
                                my_dict[index] = abs(m[1])

my_array = []
for key, value in my_dict.items():
    my_array.append([int(key),value])
my_array = np.array(my_array).T
sort_indices = np.argsort(my_array[0])
my_array[0] = my_array[0][sort_indices]
my_array[1] = my_array[1][sort_indices]
```

Calculate all coefficients

```python
pauls_gaunt = np.full(shape=(my_array[0,-1]+1,max(my_array[1]*2+1)),dtype=np.float64,fill_value=np.nan)
pauls_gaunt[0,0] = 0.0
for l1 in range(LMAX+1):
    print(l1)
    for l2 in range(LMAX+1):
        for l3 in range(abs(l1-l2), l1+l2+1):
            for m1 in range(-l1, l1+1):
                for m2 in range(-l2, l2+1):
                    m3 = -m1-m2
                    if abs(m3)<=l3:
                        l_unsorted = np.array([l1,l2,l3])
                        m_unsorted = np.array([m1,m2,m3])
                        sorted_indices = np.argsort(l_unsorted)[::-1]
                        l = l_unsorted[sorted_indices]
                        m = m_unsorted[sorted_indices]
                        index = l[0]*(6+l[0]*(11+l[0]*(6+l[0])))/24 + l[1]*(2+l[1]*(3+l[1]))/6 + l[2]*(l[2]+1)/2 + abs(m[2]) + 1
                        index = int(index)
                        if m[2]>=0:
                            my_m = m[1]
                        else:
                            my_m = -m[1]
                        index2 = (len(pauls_gaunt[index,:])-1)/2 + my_m
                        index2 = int(index2)
                        if np.isnan(pauls_gaunt[index,index2]):
                            pauls_gaunt[index,index2] = float(gaunt(l[0], l[1], l[2], m[0], m[1], m[2]).doit())
```
